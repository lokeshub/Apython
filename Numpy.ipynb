#### Operations

```python
# transpose
trans = mat.transpose()
print(trans)
```

```python
# flatten
flat = mat.flatten()
print(flat)
```

```python
print(mat)
```

```python
# ravel
raveled = mat.ravel()
print(raveled)
```

```python
print(mat)
```

###### flatten(): returns a copy of original arary. ravel() return a view( a reference) of the original array whenever possible. it does not allocate new memory unless it is  has to. 
###### because of this if you modify an element in ravel() ed array original array changes as well (because they share same memory data block)
###### ravel is faster and is just a view of the original array whenever possible

```python
raveled[0] = 99
print(raveled)
```
