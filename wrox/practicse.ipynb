```python
def odds(start = 1):
    if int(start) % 2 == 0:
        start = int(start + 1)
    while True:
        yield start
        start = start + 2
```

```python
for oddNums in odds(1):
    if oddNums > 9:
        break
    else:
        print(oddNums)
```

```python
for oN in odds(2):
    if oN > 10:
        break
    else:
        print(oN)
```
